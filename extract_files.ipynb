```python
import tarfile
import glob
import os

# 1. Define where your data lives based on your screenshot
# The folder is 'DBahn-berlin' relative to your notebooks
BASE_DIR = 'DBahn-berlin' 

def extract_archives_in_folder(subfolder_name):
    """
    Finds all .tar.gz files in a specific subfolder and extracts them 
    into that same folder.
    """
    # Construct path: DBahn-berlin/timetables/*.tar.gz
    search_pattern = os.path.join(BASE_DIR, subfolder_name, '*.tar.gz')
    archives = glob.glob(search_pattern)
    
    print(f"Found {len(archives)} archives in '{subfolder_name}'...")
    
    for archive_path in archives:
        try:
            print(f"Extracting: {archive_path}")
            # Open the tar file
            with tarfile.open(archive_path, "r:gz") as tar:
                # Extract contents to the directory containing the tar file
                extract_to = os.path.dirname(archive_path)
                tar.extractall(path=extract_to)
        except Exception as e:
            print(f"Error extracting {archive_path}: {e}")

# 2. Run extraction for both folders
# This will create the subfolders (e.g., 2509021200/) containing the XMLs
extract_archives_in_folder('timetables')
extract_archives_in_folder('timetable_changes')

print("Extraction complete.")
```
